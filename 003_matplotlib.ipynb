```python
# import the matplot lib
import matplotlib.pyplot as plt
#%matplotlib inline   #  to show the graph with in jupyter file

activity = ['online' , 'office','family','sleep','other']
hours    = [4 , 8 ,2,6,4]

plt.pie(hours ,labels=activity ,autopct='%1.3f%%')
plt.title("pie chart for daily activities")
plt.show()
```
